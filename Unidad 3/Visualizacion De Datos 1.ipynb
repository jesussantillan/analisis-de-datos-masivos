```python
import pandas as pd
import numpy as np
```

```python
df_canada=pd.read_excel('Canada.xlsx',sheet_name='Canada by Citizenship',skiprows=range(20),skipfooter=2)
```

```python
df_canada
```

```python
print(df_canada.shape)
```

```python
df_canada.drop(['AREA','REG','DEV','Type','Coverage'],axis=1,inplace=True)
```

```python
df_canada.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'},inplace=True)
```

```python
df_canada.columns=list(map(str,df_canada.columns))
```

```python
df_canada.set_index('Country',inplace=True)
df_canada.head()
```

```python
df_canada['Total']=df_canada.sum(axis=1)
df_canada.head()
```

```python
years=list(map(str,range(1980,2014)))
years
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
```

```python
df_canada.sort_values(['Total'],ascending=False,axis=0,inplace=True)
df_canada
```

```python
df_top5=df_canada.head()
df_top5
```

```python
df_top5=df_top5[years].transpose()
df_top5
```

```python
df_top5.index=df_top5.index.map(int)
df_top5
```

```python
df_top5.plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Tendencia De Imigracion')
plt.ylabel('Numero De Imigrantres')
plt.xlabel('Años')
plt.show()
```

```python
#Transparencia
df_top5.plot(kind='area',alpha=0.20,stacked=False,figsize=(20,10))
plt.title('Tendencia De Imigracion')
plt.ylabel('Numero De Imigrantres')
plt.xlabel('Años')
plt.show()
```

```python
ax=df_top5.plot(kind='area',alpha=0.35,figsize=(20,10))
ax.set_title('Tendencias de inmgracion de Canada')
ax.set_ylabel('Numero De Inmigrantes')
ax.set_xlabel('Años')
```

```python
df_chile=df_canada.loc['Chile',years]
df_chile.head()
```

```python
df_chile.plot(kind='barh',figsize=(10,6))
plt.ylabel('Años')
plt.xlabel('Numero De Imigrantes')
plt.title('Imigrantes Chilenos a canada desde 1980 hasta 2013')
for index,value in enumerate(df_chile):
    label=format(int(value),',')
    plt.annotate(label,xy=(value,index),color='black')
plt.show()
```

```python
df_mexico=df_canada.loc['Mexico',years]
df_mexico.head()
```

```python
df_mexico.plot(kind='bar',figsize=(10,6))
plt.xlabel('Años')
plt.ylabel('Numero De Imigrantes')
plt.title('Imigrantes Mexicanos a canada desde 1980 hasta 2013')
```

### Grafico Circular

```python

df_continentes=df_canada.groupby('Continen',axis=0).sum()
df_continentes
```

```python
df_continentes['Total'].plot(kind='pie',shadow=True,autopct='%1.1f%%',figsize=(6,6))
plt.title('Inmigracion por continente hacia Canada')
```

```python
explode_list=[0.1,0,0,0,0.1,0.1]
df_continentes['Total'].plot(kind='pie',shadow=None,autopct='%1.1f%%',figsize=(12,6),labels=None,pctdistance=1.15,explode=explode_list)
plt.title('Inmigracion por continente hacia Canada')
plt.legend(labels=df_continentes.index,loc='lower left')
plt.axis('equal')
```
