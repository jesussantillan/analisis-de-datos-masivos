```python
sobrevivientes=0
muertos=0
for sobrevive in datos['Survived']:
    if sobrevive == 1:
        sobrevivientes=sobrevivientes+1
    else:
        muertos=muertos+1
```

```python
y_target = datos["Survived"].values
```

```python
columns = ['RESULTADO', 'EDAD', 'DIABETES', 'EPOC', 'ASMA', 'INMUSUPR', 'HIPERTENSION']
X_input = datos[list(columns)].values
```

```python
from sklearn import tree
clf_train = tree.DecisionTreeClassifier(criterion="entropy", max_depth=7)
clf_train = clf_train.fit(X_input, y_target)
```

```python
clf_train.score(X_input,y_target)
```

```python
clasificacion = ['SOBREVIVE','NO SOBREVIVE']
```

```python
with open("covid.dot", 'w') as f:
    f = tree.export_graphviz(clf_train, out_file=f, feature_names=columns,
                          class_names=clasificacion)
```
